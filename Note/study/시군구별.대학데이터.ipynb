```python
import pandas as pd
```

```python
university = pd.read_excel("./data/대학교 수(시도:시:군:구).xlsx")
university.head()
```

```python
university = university.iloc[1:,:].replace("-",0)
university
```

```python
test = []
sid = []
sido = []
num = 1
tom = 0
for t in range(20): # 20번 
    for i in range(17): # 17
        test.append(university.columns[num]) # 년도 리스트 만드는거임.
    num+=1

for t in range(17): # 17
    sid.append(university.iloc[:,0][t])
sido.append(sid *20) # 전국 시군구를 2003~2022에 맞춰 23번 곱해 리스트를 만듦.
years = []
f = 1
for i in  range(20):
        for k in range(17):
            years.append(university.iloc[:,f][k])
        f+=1
        # university.iloc[:,1][16] # 20/16
```

### 시.도.군별 사설학원수

```python
academy = pd.read_excel("./data/사설학원수(시도:시:군:구).xlsx")
academy = academy.iloc[1:,:]
academy
```

```python
academy.head()
academy_y = academy.iloc[:,1:]
```

```python
academy_T = academy_y.T.sort_index(ascending=True).T
```

```python
academy.iloc[:,0]
```

```python
academy_T['행정구역별'] = academy.iloc[:,0]
academy_T
```

```python
academy_T.replace("-",0,inplace=True)
academy_T.info()
```

```python
academy_T.columns[19] # 0 ~ 19 
# 년도 20개 0 ~19col
# 지역 17개 
academy_T.iloc[:,16][16]
# univerStudy.iloc[:,1]
# univerStudy

len(academy_T.columns)-1
academy_T.columns[19]
academy_T.iloc[:,20][17]
academy_T.columns[20]
```

```python
test = []
sid = []
sido = []
num = 0
tom = 0
for t in range(20): # 16번 
    for i in range(17): # 17
        print(num)
        test.append(academy_T.columns[num]) # 년도 리스트 만드는거임.
    num+=1

for t in range(1,18): # 17
    sid.append(academy_T.iloc[:,20][t])
sido.append(sid *20) # 전국 시군구를 2003~2022에 맞춰 19번 곱해 리스트를 만듦.
years = []
f = 0
for i in  range(20):
        for k in range(1,18):
            years.append(academy_T.iloc[:,f][k])
        f+=1
        # university.iloc[:,1][16] # 20/16
```

```python
# row수 확인.
print(len(test))
print(len(sido[0]))
print(len(years))
```

```python
academy_T_df = pd.DataFrame(
    {'시점':test,
    '지역':sido[0],
    '사설학원수':years
    }
)
academy_T_df.head()
```

```python
academy_T_df.info()
```

```python
academy_T_df['시점'] = academy_T_df['시점'].astype('int64')
```

```python
academy_T_df.iloc[:,1].unique()
```

```python
academy_T_df.iloc[:,0].unique()
```

```python
academy_T_df.to_csv("./edudata/시군구별.사설학원_갯수변화.csv")
```
