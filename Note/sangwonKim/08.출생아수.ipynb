# 통합 파일에 신생아수 추가하기

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../Data/01.통합.csv")
df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
birth = pd.read_csv("../../Data/sangwonKim/출생아수_시도_2013_2021.csv", encoding='euc-kr')
birth.columns = ['년도', '시도', '출생아수']
birth
```

```python
df_2 = pd.merge(df, birth, how='left', on=['년도', '시도'])
df_2
```

```python
df_final = df_2[['년도','시도','총인구','전출인구','출생아수','산부인과의원수','초등학생인원수','빈집수','협회 및 단체, 수리 및 기타 개인서비스업 소멸','도매 및 소매업 신생','교육서비스업 소멸','예금은행대출금액','예금액']]
df_final
```

```python
df_final.to_csv("../../Data/02.통합.csv")
```
