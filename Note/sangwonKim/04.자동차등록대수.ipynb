## 시도별 자동차 등록대수

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../Data/sangwonKim/자동차등록대수현황_시도별_20230208.csv", encoding='euc-kr')
df.head()
```

```python
# 필요없는 행 삭제하기
df.drop(columns=['레벨01(1)', '항목'], inplace=True)
df.drop(index=0, axis=0, inplace=True)
df.head()
```

```python
df.rename(columns={'시점':'연도'}, inplace=True)
df.head()
```

```python
# 연도 리스트 구하기
year_list = []
for i in range(2013, 2023):
    year_list += [i] * 12

year_list
```

```python
df.연도 = year_list
df.head()
```

```python
# 결측치 확인
df.isna().sum()
```

```python
df.info()
```

```python
# data: object -> int 로 변경하기
df = df.astype(int)
df.info()
```

```python
df.연도 = df.연도.astype(object)
df.info()
```

```python
df.columns[1:]
```

```python
df = pd.pivot_table(data=df, index='연도', sort=False , aggfunc='sum')
df.head()
```
