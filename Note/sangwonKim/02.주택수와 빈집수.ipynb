### 2015 ~ 2021 년 까지의 빈집수, 주택수 가져오기

```python
# 빈집수 가져오기
house_empty = []
for i in range(2015, 2022):
    house_empty += list(df[f"{i}.1"])

house_empty
```

```python
# 주택수 가져오기
house = []
for i in range(2015, 2022):
    house += list(df[f"{i}.2"])

house
```

```python
df.head()
```

```python
# 년도 배열
years = []
for i in range(2015, 2022):
    years = years + [i]*17

years
```

```python
# 시도 배열
sido = list(df.index)
sido = sido*7
sido
```

```python
# 리스트 길이 확인
len(years) == len(sido)
```

```python
# dataframe 틀 만들기
df_house = pd.DataFrame({'년도':years, '시도':sido})
df_house.set_index('년도', inplace=True)
df_house
```

## 데이터프레임 컬럼 추가하기

```python
df_house['주택수'] = house
df_house['빈집수'] = house_empty
df_house.head()
```

## 인구 데이터프레임과 주택/빈집수 데이터프레임 합치기

```python
df_merge = pd.merge(df_pop, df_house, how='left', on=['년도', '시도'])
df_merge
```
