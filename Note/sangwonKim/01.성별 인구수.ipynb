### ___.1,2,3 가져오기

```python
# 전체 인구수 가져오기
pop_all = []
for i in range(2013, 2023):
    pop_all += list(df2[f"{i}"])

pop_all
```

```python
# 남자 인구수 가져오기
pop_male = []
for i in range(2013, 2023):
    pop_male += list(df2[f"{i}.1"])

pop_male
```

```python
# 여자 인구수 가져오기
pop_female = []
for i in range(2013, 2023):
    pop_female += list(df2[f"{i}.2"])

pop_female
```

```python
# 컬럼 추가하기
df_pop['총 인구'] = pop_all
df_pop['남성 인구'] = pop_male
df_pop['여성 인구'] = pop_female
df_pop
```

```python
df_pop.set_index('년도', inplace=True)
df_pop
```

```python
df_pop.info()
```

```python
df_pop.iloc[:,1:] = df_pop.iloc[:,1:].astype(dtype=int)
df_pop.info()
```
