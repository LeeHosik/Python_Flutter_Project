### ___.1,2,3 가져오기

```python
# 전체 인구수 가져오기
pop_all = []
for i in range(2013, 2023):
    pop_all += list(df2[f"{i}"])

pop_all
```

```python
# 남자 인구수 가져오기
pop_male = []
for i in range(2013, 2023):
    pop_male += list(df2[f"{i}.1"])

pop_male
```

```python
# 여자 인구수 가져오기
pop_female = []
for i in range(2013, 2023):
    pop_female += list(df2[f"{i}.2"])

pop_female
```

```python
# 컬럼 추가하기
df_pop['총 인구'] = pop_all
df_pop['남성 인구'] = pop_male
df_pop['여성 인구'] = pop_female
df_pop
```

```python
df_pop.set_index('년도', inplace=True)
df_pop
```

```python
df_pop.info()
```

```python
df_pop.iloc[:,1:] = df_pop.iloc[:,1:].astype(dtype=int)
df_pop.info()
```

# 각 시도별로 데이터 뽑아내서 리스트에 담기

```python
# 나중에 코드 재사용하기
pop_sido = []
for sido in sido_list:
    pop_sido.append(df_pop[df_pop['시도']==sido]['총 인구'])

pop_sido
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# plt.style.use('ggplot')

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(1,1,1)

for sido in pop_sido:
    ax.plot(
        sido.index,
        sido.values/1000,
        marker = 'o',
        markersize = 5,
    )

ax.legend(labels=sido_list, loc='best')
ax.set_title('각 시도별 인구', size=20)
ax.set_xlabel('기간', size=12)
ax.set_ylabel('총 인구(단위:1000명)', size=12)
ax.set_xticklabels(sido_list, rotation=0)
# ax.tick_params(axis='x', labelsize=10) # 해도그만 안해도그만

plt.show()
```
