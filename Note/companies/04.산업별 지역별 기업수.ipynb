> 지역별로 업종을 합치고 년도별로 쌓기

```python
# 년도 데이터프레임
sido = pd.DataFrame(sido_arr)

sido_copy = sido.copy()
sido_copy
sido_copy['년도'] = 2013
df = sido_copy
years = range(2013, 2021)

for year in years:
    sido_copy = sido.copy()
    sido_copy['년도'] = year
    df = pd.concat([df, sido_copy], axis=0)
    
df.reset_index(drop=True, inplace=True)
df.columns = ['시도', '년도']
df
```
