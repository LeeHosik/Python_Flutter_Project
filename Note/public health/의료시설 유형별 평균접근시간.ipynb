```python
# years = [2017, 2018, 2019, 2020]
years = [2019, 2020]
colors = ['green','palevioletred','blue','navy','red','skyblue','coral','darkviolet','mediumspringgreen','navy','yellow','pink','darkgreen','palevioletred']
# fig = plt.figure(figsize=(25,10))

for year in years:
    titleYear = year
    ax1 = pd.pivot_table(goHospital[goHospital.index == year], index = '시도별', aggfunc = 'mean').plot(
        kind='bar',
        figsize=(20,10),
    )
    ax2 = ax1.twinx()
    ax2.plot(
        go20ko,
        marker="o",
        markerfacecolor="red",
        markersize=10,
        # color="pink",
        linewidth=2,
        # label="서울 -> 부산"
    )

    ax1.set_ylim(0, 80)
    ax2.set_ylim(0, 80)
    plt.xticks(rotation="horizontal",size=10) # vertical은 90도
    plt.ylabel("소요시간(단위:분)", size=15)
    plt.title(' %d 년 의료시설 유형별 일 평균접근시간'%titleYear)
    ax1.legend(loc="upper right")
    ax2.legend(labels=go2020.columns ,loc="upper right")
    plt.show()
```
