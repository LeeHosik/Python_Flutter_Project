## k-means -> 년도별 최적의 k값 찾기

```python
# plt.style.use('ggplot')
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(1,1,1)

for year in range(2013, 2021):
    df_temp = df[df.index == year]
    df_temp.set_index('시도', inplace=True)
    df_km = df_temp[df_temp.columns.difference(['총인구'])]
    # k값 찾기 --------------------------------------------
    scores = []
    for k in range(2, 10):
        km = KMeans(n_clusters=k, random_state=42)
        km.fit(df_km)
        scores.append(km.inertia_)
    # 그래프 그리기 ------------------
    ax.plot(range(2,10), scores, label = year, lw=1, marker='o', markersize = 5)

ax.set_title("년도별 k값 찾기")
ax.legend(loc='best')
ax.set_xlabel("k 값")


# ------------
# 주석 표시 - 화살표 (annotation: )
plt.annotate(
    '',
    xy=(2.3, 9e11), # 화살표의 머리 부분(끝점)
    xytext=(2.3, 4e11), # 화살표의 꼬리 부분(시작점)
    xycoords='data', # 좌표체계     coordinates
    arrowprops=dict(arrowstyle='->', color='blue', lw=5) # 화살표 서식    '-' , '->'
)

# 주석 표시 - 화살표 (annotation: )
plt.annotate(
    '',
    xy=(4.2, 4.0e11), # 화살표의 머리 부분(끝점)
    xytext=(3.1, 3.1e11), # 화살표의 꼬리 부분(시작점)
    xycoords='data', # 좌표체계     coordinates
    arrowprops=dict(arrowstyle='->', color='red', lw=2) # 화살표 서식    '-' , '->'
)

# 주석 표시 - 텍스트
plt.annotate(
    'k값(3)의 가능성 상승(2013~2020)',
    xy=(6.0, 4e11),
    rotation=0,
    va='baseline', # 상하정렬: vertical alignment
    ha='center', # 좌우정렬
    fontsize=12
)
# ------------

plt.show()
# -----------------------------
```

## 전체 년도의 k값 찾기

```python
df_temp = df.dropna(axis=0)
df_temp.set_index('시도', inplace=True)
df_km = df_temp[df_temp.columns.difference(['총인구'])]
# k값 찾기 --------------------------------------------
scores = []
for k in range(2, 10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(df_km)
    scores.append(km.inertia_)
# 그래프 그리기 ------------------
plt.figure(figsize=(8,5))
plt.plot(range(2,10), scores, marker='o', markersize = 5)
plt.title('전체 년도의 k값 찾기')
plt.show()
# -----------------------------
```
