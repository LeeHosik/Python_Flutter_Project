# 서울 데이터 수집

```python
import pandas as pd
import numpy as np
```

```python
df_inout = pd.read_csv("../Data/전입전출인구_서울_군구별_2013-2022.csv", encoding='euc-kr')
df_inout.rename(columns={'행정구역(시군구)별':'시군구'}, inplace=True)
df_inout
```

```python
# df = df_inout.copy()
# df.rename(columns={'행정구역(시군구)별':'시군구'}, inplace=True)
# df
```

```python
df_pop = pd.read_csv("../Data/주민등록인구_등록외국인수_서울_군구별_2013-2021.csv", encoding="euc-kr")
df_pop.drop(columns='행정구역별(1)', axis=1, inplace=True)
df_pop.rename(columns={'행정구역별(2)':'시군구'}, inplace=True)
df_pop
```

```python
df_birth = pd.read_csv("../Data/출생아수_서울_군구별_2013-2022.csv", encoding="euc-kr")
df_birth.rename(columns={'시군구별':'시군구', '계 (명)':'출생아수'}, inplace=True)
df_birth
```

```python
df_hospital = pd.read_csv("../Data/산부인과_소아청소년과_서울_군구별_2013-2022.csv", encoding="euc-kr")
df_hospital.drop(columns='시군구별(1)', axis=1, inplace=True)
df_hospital.rename(columns={'시군구별(2)':'시군구', '산부인과':'산부인과의원수', '소아청소년과':'소아청소년과의원수'}, inplace=True)
df_hospital
```

```python
temp_list = []
for i in range(0, 250):
    temp = df_hospital['시점'][i].replace('.4/4','')
    temp_list.append(temp)
df_hospital['시점'] = temp_list
df_hospital['시점'] = df_hospital['시점'].astype(int)
df_hospital
```

```python
df_culture = pd.read_csv("../Data/문화기반시설수_서울_군구별_2013-2021.csv", encoding="euc-kr")
df_culture.rename(columns={'행정구역별':'시군구', '문화기반시설수<br>(A) (개)': '문화기반시설수'}, inplace=True)
df_culture
```

```python
df = pd.merge(df_inout, df_pop, on=['시군구', '시점'])
df = pd.merge(df, df_birth, on=['시군구', '시점'])
df = pd.merge(df, df_hospital, on=['시군구', '시점'])
df = pd.merge(df, df_culture, on=['시군구', '시점'])
df
```

```python
df.columns = ['시군구','년도','전입인구','전출인구','등록외국인','주민등록인구','출생아수','산부인과의원수','소아청소년과의원수','문화기반시설수']
df.head()
```
